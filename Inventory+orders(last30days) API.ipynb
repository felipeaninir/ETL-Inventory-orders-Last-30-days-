```python
inventory_df = []
for account_id in clientes.keys():
    key = clientes[account_id]
    payload = {
      'key': key,
      'warehouse': 'bx1',
      'account_id': account_id,
      'page': '1',
      'Content-Type': 'text/plain'
    }
    
    response = requests.request("POST", url, headers=headers, data=json.dumps(payload), verify=False)

    json_obj = response.json()
    total_pages = int(json_obj['total_pages'])
    
    print(f'Account ID: {account_id}')
    print(f'Pages: {total_pages}')
    print("-------------------")
    
    for npage in range(total_pages+1):
        payload = {
          'key': key,
          'warehouse': 'bx1',
          'account_id': account_id,
          'page': npage,
          'Content-Type': 'text/plain'
        } 

        response = requests.request("POST", url, headers=headers, data=json.dumps(payload), verify=False) 
        json_obj = response.json()
        pages_df = pd.json_normalize(json_obj, record_path=['products'])

        inventory_df.append(pages_df)
        
inventory = pd.concat(inventory_df, ignore_index=True)
print(inventory.shape)
```

```python
url = "https://travelwifiintegration.shipedge.com/API/Rest/v2/Orders//get_orders"

headers = {
  'key': '7c1277b835a0fc4873b90174afd1e132',
  'warehouse': 'bx1',
  'account_id': '35',
  'page': '1',
  'status': 'all',
  'Content-Type': 'text/plain'
}

today = datetime.today().strftime('%Y-%m-%d')
last30days = (pd.to_datetime(today) - timedelta(days=30)).strftime('%Y-%m-%d')

orders_df = []
for account_id in clientes.keys():
    key = clientes[account_id]
    payload = {
      'key': key,
      'warehouse': 'bx1',
      'account_id': account_id,
      'page': '1',
      'status': 'all',
      'from_date': last30days,
      'Content-Type': 'text/plain'
    }
    
    response_Orders = requests.request("POST", url, headers=headers, data=json.dumps(payload), verify=False)

    json_obj_Orders = response_Orders.json()
    total_orders = int(json_obj_Orders['total_orders'])
    total_pages = int(json_obj_Orders['total_pages'])
    
    print(f'Account ID: {account_id}')
    print(f'Pages: {total_pages}')
    print(f'Orders: {total_orders}')
    print("-------------------")
    
    for npage in range(total_pages+1):
        payload = {
          'key': key,
          'warehouse': 'bx1',
          'account_id': account_id,
          'page': npage,
          'status': 'all',
          'from_date': last30days,
          'Content-Type': 'text/plain'
        } 

        response = requests.request("POST", url, headers=headers, data=json.dumps(payload), verify=False) 
        json_obj = response.json()
        
        for order in range(len(json_obj['orders'])):
            if json_obj['orders'][order]['status']['description']=='shipped':
                #orders_json.append(json_obj['orders'][order])
        
                pages_df = pd.json_normalize(json_obj['orders'][order], record_path=['items'], meta=['order_id', 'date', 'pending_date', 'shipping_date', 'last_update', ['shipping', 'tracking_number'], ['account', 'account_id'], ['account', 'account_name'], ['account', 'company_name'], ['status', 'description']])

                orders_df.append(pages_df)
        
orders = pd.concat(orders_df, ignore_index=True)

print(orders.shape)
print(len(orders.order_id.unique()))
print("---------------") 
print(orders.shape)
print(f'ordenes: {len(orders.order_id.unique())}')
```
